```python
import cv2
```

```python
# capture
cap = cv2.VideoCapture(0) # 0, default kamera. farklı kameralarımız varsa 0, 1 falan yazarız

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(width, height)  # 640 yatay, 480 dikey piksel boyutu var
```

```python
# record video
writer = cv2.VideoWriter("my_video.mp4", cv2.VideoWriter_fourcc(*"DIVX"), 20, (width, height))
# my_video.mp4 -> videomuzun adı
# cv2.VideoWriter_fourcc(*"DIVX") -> windows için: DIVX, frameleri sıkıştırmak için kullanılan 4 karakterli codec kodu
# 20 -> fps (frames per second)
# width and height of the frame
```

```python
# read the video
while True:
    ret, frame = cap.read()
    cv2.imshow("frame", frame)

    # save
    writer.write(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
cap.release()  # kameranın bağlantısını keser
writer.release()  # video dosyasını kaydeder ve kapatır
# birbirlerinden bağımsız kaynakları serbest bıraktıkları için yer değiştirseler sorun yok
cv2.destroyAllWindows()
```
