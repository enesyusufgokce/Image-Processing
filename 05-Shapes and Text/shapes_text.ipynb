```python
import cv2
import numpy as np
```

```python
# create image
img = np.zeros((512,512,3), np.uint8) # 0-255 e kadar oaln değerler arasında integerler alır. ve sıfırlar int olcak
# siyah bir resme karşılık geliyor çünkü pixelin genliği sıfırken rengi siyah idi. 1 desek beyaz olurdu
print(img.shape)
cv2.imshow('black', img)

cv2.waitKey(0)
cv2.destroyAllWindows() # q ya değil, direkt çarpıya basınca çıktık. q lularda da direkt çarpıyla çıkabilirsin
```

```python
# line
# image, başlangıç noktası, bitiş noktası, renk, thickness
cv2.line(img, (0,0), (512,512), (32,143,134), 3)
# opencv renk kodunu rgb değil de bgr olarak kabul ediyo

cv2.imshow('line',img)

cv2.waitKey(0)
cv2.destroyAllWindows()

"""
# En dıştaki [  -> Resmin başlangıcı
[
  [[B,G,R], [B,G,R]],  # 1. satır
  [[B,G,R], [B,G,R]]   # 2. satır
]
# En dıştaki ]  -> Resmin bitişi
"""
```

```python
# rectangle
cv2.rectangle(img, (0,0), (256,256), (255,0,0), cv2.FILLED)
cv2.imshow("rectangle:", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
.# circle
# image, center, radius, color
cv2.circle(img, (133,122), 32, (111,145,65), cv2.FILLED)
cv2.imshow("circle:", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
