```python
import cv2
import time
```

```python
# video name
video_name = "MOT17-04-DPM.mp4"

# import video: capture, cap
cap = cv2.VideoCapture(video_name)

print("Width:", cap.get(3))  # get in 3. indeksinde genişlik
print("Height", cap.get(4))  # 4. indexinde boy temsil edilir
```

```python
if not cap.isOpened():
    print("Error opening video stream or file")  # .mp4 ü unuttuğumda hatayı veriyor
```

```python
# read the video
while True:
    ret, frame = cap.read() # frame, videonun içinde bulunan her bir resim, return ise read işleminin başarılı olup olamadığı. cap.read() başarısızsa return false döncek

    if ret:
        time.sleep(0.01)  # çok hızlı olursa anlaayamayabiliriz
        cv2.imshow("Frame", frame) # resmi okuyoruz. bunun video haline gelmesi için bu read işlemini art arda yapmamız lazım (while)
    else: break  # videoda artık gösterilecek bir şey kalmamıştır

    if cv2.waitKey(1) & 0xFF == ord("q"):  # waitkey 1 demek, frame nin çok kısa süre durup hemen yoluna devam edeceği anlamına gelir(videolar zaten framelerin art arda çok hızlı geçmeleriyle oluşur)
        break  # ben kendi isteğimle çıktım demektir

cap.release() # stop capture
cv2.destroyAllWindows()  # açık tüm pencereleri kapat
```
