```python
# gradients, görüntüdeki yoğunluk veya renkteki yönlü bir değişikliktir
# edge detection
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sudoku.jpg", 0)
plt.figure, plt.imshow(img, cmap="gray"), plt.axis("off"), plt.title("Original Image"), plt.show()
```

```python
# x gradient (türev) dx=1 dy=0 de sütunlar, x ekseni boyunca ilerlerken rekn değiştirdiği için fark edilirler
# ddepth, outputun derinliğini gösterir. ksize=5 (5*5 kil kutucuk)
sobelx = cv2.Sobel(img, ddepth=cv2.CV_16S, dx=1, dy=0, ksize=5)
plt.figure, plt.imshow(sobelx, cmap="gray"), plt.axis("off"), plt.title("Sobel X"), plt.show()
```

```python
# y gradient
sobely = cv2.Sobel(img, ddepth=cv2.CV_16S, dx=0, dy=1, ksize=5)
plt.figure, plt.imshow(sobely, cmap="gray"), plt.axis("off"), plt.title("Sobel Y"), plt.show()
```

```python
# laplacian gradient
laplacian = cv2.Laplacian(img, ddepth = cv2.CV_16S)
plt.figure, plt.imshow(laplacian, cmap="gray"), plt.axis("off"), plt.title("Laplace"), plt.show()
```
