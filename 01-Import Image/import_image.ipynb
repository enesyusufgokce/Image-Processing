NOTE:
- genlik (Amplitude), pixelin sahip olduğu değerin büyüklüğü. 0 ise siyah, 255 ise beyaz

```python
import cv2
```

```python
# import
img = cv2.imread("messi.jpg", 0)  # 0, resmi grayscale olarak import etmemizi sağlıyo
```

```python
# visualize
cv2.imshow("first picture", img)
k = cv2.waitKey(0) & 0xFF  # fotoğraf için 1 dersek bir pencere saliseler içinde açılıp anında kapanır

if k == 27:  # 27, esc nin klavyedeki karşılığı
    cv2.destroyAllWindows()  # açık olan tüm windowları kapat
elif k == ord("s"):  # s ye basınca:
     cv2.imwrite("messi_gray.png", img)  # png olarak kaydet (esc ye basarsak kapanır, s ye basarsak kaydeder)
     cv2.destroyAllWindows()
```
