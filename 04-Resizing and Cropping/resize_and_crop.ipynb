```python
import cv2
```

```python
img = cv2.imread('lenna.png')
print('shape of the picture:', img.shape)
# NOTE: imread de 1 yerine 0 dersek shape, (512, 512) olur. 1 iken renkli olduğu için RGB den 3 geldi. (normalde görünmeyen 1 vardı(tek renk))

cv2.imshow("original", img)
```

```python
# resized
img_resized = cv2.resize(img, (800,800))
print("resized img shape:", img_resized.shape)
cv2.imshow("resized img", img_resized)
```

```python
# crop
img_cropped = img[:200,:300]  # height(y axis), width(x axis)
cv2.imshow("cropped img: ", img_cropped)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
