```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("img1.JPG")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) # opencv bgr okuduğu için rgb ye çeviriyoz, orijinal img için
img2 = cv2.imread("img2.JPG")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img1, vmin=0, vmax=255)
plt.axis("off")
plt.show()

plt.figure()
plt.imshow(img2, vmin=0, vmax=255)
plt.axis("off")
plt.show()
```

```python
# birleştirmek için aynı boyutta olmak zorundalar

print(img1.shape)
print(img2.shape)

img1 = cv2.resize(img1, (600,600))
img2 = cv2.resize(img2, (600,600))
print(img1.shape, img2.shape)

plt.figure()
plt.imshow(img1, vmin=0, vmax=255)
plt.axis("off")
plt.show()

plt.figure()
plt.imshow(img2, vmin=0, vmax=255)
plt.axis("off")
plt.show()
```

```python
# mixed img = alpha*img1 + beta*img2
blended = cv2.addWeighted(src1=img1, src2=img2, alpha=0.5, beta=0.7, gamma=0)

plt.figure()
plt.imshow(blended, vmin=0, vmax=255)
plt.axis("off")
plt.show()
```
